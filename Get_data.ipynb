```python
import Player
import Game
import pandas as pd
import numpy as np
import time
```

```python
data = {
   "turn number" : [],
    "Max number" : [],
    "Max adjacent": [],
    "Time spent" : [],
    "parameter_second" : [],
    "parameter_turn_weight" : [],
    "table size" : []
}
```

```python
def get_data(second=0.5, weight=0.01, table_size=5):
    player1 = Player.MonteCarloSecond(second=second)
    game = Game.Game(player1, table_size=table_size)
    start_time = time.time()
    result = game.play()
    time_spent = time.time() - start_time
    data["turn number"].append(result[1])
    data["Max number"].append(result[2])
    data["Max adjacent"].append(result[3])
    data["Time spent"].append(time_spent)
    data["parameter_second"].append(player1.parameter)
    data["parameter_turn_weight"].append(player1.turn_weight)
    data["table size"].append(game.table_size)
```

```python
second_list = np.linspace(0.1, 1.0, 4)
```

```python
for second in second_list:
    for i in range(3):
        get_data(second=second)
```

```python
df_data = pd.DataFrame(data)
df_data
```

```python
full_data = pd.read_csv("data.csv")
full_data.append(df_data, ignore_index=True)
```

```python
full_data.to_csv("data.csv", index=False)
```
